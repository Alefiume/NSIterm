Faire une boucle

Soit on sait directement combien de fois il faut réaliser une opération. on utilise donc for, soit on a une condition pour arrêter la boucle: on utilise while

```python
for valeur in range(1,6,2):# valeur est un nom de variable 
    print(valeur) # On peut donc l'utiliser ici
```

Range prend 3 arguments: Le premier est la valeur initiale le deuxième est kla première valeur nion atteinte par le troisième est le pas.

Donc ici on va de 1 à 6 (exclu) par pas de 2 c'est a dire 



```python
for i in range ( 0,6,1):
    print(i)
```

```python
for _  in range (6):
    print(_)
```

```python
i = 0
while i < 10:
    print(i)
    i += 1 # identique  à i = i + 1
print("fin", i)
```

Question 4:
    Ecrire une fonction qui calcule et renvoie la valeur de n! = n *(n-1)*(n-2)*...*2*1

```python
def factorielle(x):
    rep = 1
    for i in range (1, x+1):
        rep= i * rep
    return rep
factorielle(7)
```
